```python
# task 1
# reverse string

def reverse(text: str) -> str:
    reversed_str: str = ''
    for item in text:
        reversed_str = item + reversed_str

    return reversed_str

reverse('Hello world!')
```

```python
# task 2
# count vowels and consonant

def count_letters(text: str) -> dict:
    vowel: str = 'aeiou'
    returned_dict: dict = {
        'vowel': 0,
        'consonant': 0
    }

    text: str = ''.join([item for item in text.lower() if item.isalpha()])

    for item in text:
        if item in vowel:
            returned_dict['vowel'] += 1
        else:
            returned_dict['consonant'] += 1

    return returned_dict

count_letters('Test this string')
```

```python
# task 3
# palindrom

def is_palindrom(text: str) -> bool:
    text: str = ''.join([item for item in text.lower() if item.isalpha()])
    return text == reverse(text)
    
is_palindrom('А роза упала на лапу Азора')
```

```python
# task 4
# delete duplicates

def delete_duplicates(text: str) -> str:
    dictionary: dict = {}
    res: str = ''

    for item in text:
        if item not in dictionary:
            dictionary[item] = True
            res+= item

    return res

delete_duplicates('aasseerfett')
```

```python
# task 5
# words count

def words_count(text: str) -> int:
    return len(text.split(' '))

words_count('I want to count this words!')
```

```python
# task 6
# Caesar cipher

def caesar_cipher(text: str, shift: int) -> str:
    alphabet = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    res = ''

    for item in text:
        if item in alphabet:
            new_index = alphabet.find(item) + shift
            res += alphabet[new_index]
        else:
            res += item
    
    return res

caesar_cipher('Caesar cipher!', 2)
```

```python
# task 7
# capitalized first letters

def capitalized_first_letters(text: str)-> str:
    res: list = []
    words_list: list = text.split(' ')
    
    for word in words_list:
        word = word[0].upper() + word[1:]
        res.append(word)

    return ' '.join(res)

capitalized_first_letters('i want to capitalized all words')
        
```

```python
# task 8
# longest uniq substr

def longest_uniq_substr(text: str) -> str:
    dictionary: dict = {}
    res: str = ''
    left: int = 0

    for i, char in enumerate(text):
        if char in dictionary:
            left = max(left, dictionary[char] + 1)

        dictionary[char] = i
        current = text[left:i + 1]
        
        if len(res) < len(current): 
            res = current

    return res

longest_uniq_substr('zxcvbnzxcvghfgfqwertyuiop')
```

```python
# task 9
# braces

def braces_validation(braces: str) -> bool:
    pairs = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    stack = []

    for brace in braces:
        if brace in pairs.values():
            stack.append(brace)
        else:
            if not stack:
                return False

            if stack[-1] != pairs[brace]:
                return False

            stack.pop()

    return len(stack) == 0

braces_validation('({{[()]}})')
```

```python
# alternation

def alternation(list_1: list, list_2: list) -> list:
    res: list = []
    shortest_list = list_1 if len(list_1) < len(list_2) else list_2
    longest_list = list_2 if len(list_1) < len(list_2) else list_1

    for a, b in zip(list_1, list_2):
        res.extend([a, b])

    res.extend(longest_list[len(shortest_list):])

    return res

alternation([1, 2, 3], [4, 5, 6, 7, 8, 9])
```

```python
# transfer zeros

def transfer_zeros(array: list) -> list:
    zeros_count: int = 0
    res: list = []

    for item in array:
        if item == 0:
            zeros_count += 1
        else:
            res.append(item)
    
    for _ in range(zeros_count):
        res.append(0)

    return res

transfer_zeros([0, 1, 0, 0, 3])
```

```python
# max mark

def max_value(diary: dict) -> str:
    max_val:int = 0

    for value in diary.values():
        if value > max_val:
            max_val = value

    for name, mark in diary.items():
        if mark == max_val:
            return name


max_value({'Аня': 85, "Боря": 92, "Витя": 78})
```

```python
# hash tower

def hash_tower(n: int) -> str:

    for str_number in range (1, n + 1):
        spaces = n - str_number
        hashes: int = str_number * 2 - 1
        print(" " * spaces + "#" * hashes)

hash_tower(4)
```
