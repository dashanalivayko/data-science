```python
# task 1

import math

x: float = 1.79

y = (
    math.cos(math.e ** x)
    + math.log(1 + x) ** 2
    + math.sqrt(
        math.e ** math.cos(x)
        + math.sin(math.pi * x) ** 2
    )
    + math.sqrt(1 / x)
    + math.cos(x ** 2)
) ** math.sin(x)

y
```

```python
# task 2

def search_substr(subst: str, st: str) -> str:
    return 'Есть контакт!' if st.lower().find(subst.lower()) >= 0 else 'Мимо!'

search_substr("bro", "Hej, what's up Bro?")
```

```python
# task 3

from collections import Counter

def counter(str: str) -> list:
    return Counter(str.replace(' ', '')).most_common(3)

counter('Returns 0 for missing keys, whereas a dictionary raises a KeyError')
```

```python
# task 4

from collections import Counter

def count_it(sequence: str) -> dict:
    count = Counter(sequence).most_common(3)
    return dict(count)

count_it('1892756343845742')
```

```python
# task 5.1

def chess_queen(a: int, b: int, c: int, d: int) -> bool:
    return a == c or b == d or abs(a - c) == abs(b - d)

chess_queen(1,2,1,6)
```

```python
# task 5.2

def chess_horse(a: int, b: int, c: int, d: int) -> bool:
    return abs(a - c) * abs(b - d) == 2 

chess_horse(1,2,2,4)
```

```python
# task 5.3

def chess_horse_2(a: int, b: int, c: int, d: int) -> bool:
    steps = (1, 2, -1, -2)

    if chess_horse(a, b, c, d):
        return True
        
    return any(
        abs(dx * dy) == 2
        and 1 <= a + dx <= 8 
        and 1 <= b + dy <= 8 
        and chess_horse(a + dx, b + dy, c, d)
        for dx in steps
        for dy in steps
    )

chess_horse_2(2,3,4,4)
```
