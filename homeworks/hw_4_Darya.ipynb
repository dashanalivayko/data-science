```python
# winners tournament

def winners_tournament(lst: list) -> str:
    max_int: int = 0
    attemps: int = 0

    for i in range(len(lst) - 1):
        max_int = max(lst[i], lst[i+1])
        attemps += 1
    return f'max value: {max_int}, attemps: {attemps}'

winners_tournament([1, 5, 7, 2, 9, 2, 10])
```
