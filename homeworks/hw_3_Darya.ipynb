```python
# 1

def from_string_to_list(s: str) -> list:
    return s.split(' ')

from_string_to_list('From string to list')
```

```python
# 2

def litres(time: int) -> int:
    water = 0.5
    return round(time * water)

litres(5)
```

```python
# 3

def create_reverse_list(n: int) -> list:
    return list(range(n, 0, -1))

create_reverse_list(8)
```

```python
# 4

def century(year: int) -> int:
    return (year + 99) // 100
        
century(1736)
```

```python
# 5

def list_from_number(number: int) -> list:
    return list(str(number)[::-1])

list_from_number(12345)
```

```python
# 6

def vertical_snake(n):
    matrix = [[0]*n for _ in range(n)]
    num = 1

    for j in range(n):
        if j % 2 == 0:
            for i in range(n):
                matrix[i][j] = num
                num += 1
        else:
            for i in range(n-1, -1, -1):
                matrix[i][j] = num
                num += 1

    return matrix

vertical_snake(5)

def horizontal_snake(n):
    matrix = [[0]*n for _ in range(n)]
    num = 1

    for j in range(n):
        if j % 2 == 0:
            for i in range(n):
                matrix[j][i] = num
                num += 1
        else:
            for i in range(n-1, -1, -1):
                matrix[j][i] = num
                num += 1

    return matrix

horizontal_snake(8)
```
