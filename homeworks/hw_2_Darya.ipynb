```python
import math

def solve_quadratic(a: int, b: int, c: int) -> tuple:
    D = b**2 - 4*a*c

    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)
        return x1, x2
    
    elif D == 0:
        x = -b / (2*a)
        return (x,)
    
    else:
        return None

def solve_quadratic_with_changing(a: int, b: int, c: int) -> tuple:
    result = []
    t_roots = solve_quadratic(a, b, c)

    if t_roots is None:
        return tuple(result)

    for t in t_roots:
        x2 = t + 8

        if x2 >= 0:
            result.append(math.sqrt(x2))
            result.append(-math.sqrt(x2))

    return tuple(result)

# solve_quadratic(1, -12, 20)
# solve_quadratic(1, 17, 72)
# solve_quadratic(1, -7, -44)
# solve_quadratic(1, 9, 8)
# solve_quadratic(1, -2, -63)
solve_quadratic_with_changing(1, 4, -5)
```

```python
# reverse list

def reverse_list(array: list) -> list:
    n = len(array)

    for i in range(n // 2):
        array[i], array[-1 - i] = array[-1 - i], array[i]

    return array

reverse_list([1, 2, 3, 4, 5])
```

```python
# change list

def change_list(array: list) -> list:

    array[0], array[-1] = array[-1], array[0]

    return array

change_list([1, 2, 3, 4, 5])
```

```python
# useless int

def useless_int(array: list) -> int:
    max_int: int = 0

    for item in array:
        max_int = max(max_int, item)

    return max_int/(len(array))

useless_int([1, 2, 3, 4, 5])
```

```python
# equal string len

def all_eq(lst: list) -> list:
    max_len = max(len(item) for item in lst)

    return [item + '_' * (max_len - len(item)) for item in lst]

all_eq(['orange', 'banana', 'meat', 'tea'])
```
