```python
from abc import ABC, abstractmethod

class Salary(ABC):
    @property
    @abstractmethod
    def salary(self) -> float | int:
        pass

class Employee(Salary):
    def __init__(self, salary: float | int):
        self._salary = salary

    @property
    def salary(self) -> float | int:
        return self._salary

class SalaryDecorator(Salary):
    def __init__(self, employee: Salary):
        self.employee = employee

    @property
    def salary(self) -> float | int:
        return self.employee.salary

class PerformanceBonus(SalaryDecorator):
    def __init__(self, employee: Salary, performance_score: float):
        super().__init__(employee)
        self.performance_score = performance_score

    @property
    def salary(self) -> float:
        return super().salary + self.performance_score * 100

class SeniorityBonus(SalaryDecorator):
    def __init__(self, employee: Salary, years: int):
        super().__init__(employee)
        self.years = years

    @property
    def salary(self) -> float:
        return super().salary + self.years * 50

class ManagementBonus(SalaryDecorator):
    @property
    def salary(self) -> float:
        return super().salary + 1000
```

```python
employee = Employee(3000)

employee = PerformanceBonus(employee, 500)
employee = SeniorityBonus(employee, 3)
employee = ManagementBonus(employee)

print(employee.salary)
```
