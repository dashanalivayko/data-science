```python
from abc import ABC, abstractmethod

class Transport(ABC):
    @abstractmethod
    def deliver(self):
        pass

class Truck(Transport):
    def deliver(self):
        print("Truck is delivering goods by road.")

class Ship(Transport):
    def deliver(self):
        print("Ship is delivering goods by sea.")

class Plane(Transport):
    def deliver(self):
        print("Plane is delivering goods by air.")

class Logistics:
    def create_transport(self, transport_type: str) -> Transport:
        if transport_type == "truck":
            return Truck()

        if transport_type == "ship":
            return Ship()

        if transport_type == "plane":
            return Plane()

        raise ValueError("Unknown transport type")

    def plan_delivery(self, transport_type: str):
        transport = self.create_transport(transport_type)
        transport.deliver()
```

```python
logistics = Logistics()

logistics.plan_delivery("truck")
logistics.plan_delivery("ship")
logistics.plan_delivery("plane")
```
