```python
from abc import ABC, abstractmethod

class Computer:
    def __init__(self) -> None:
        self.cpu: str | None = None
        self.storage: str | None = None
        self.motherboard: str | None = None

    def __str__(self) -> str:
        return (
            f"CPU: {self.cpu}, "
            f"Storage: {self.storage}, "
            f"Motherboard: {self.motherboard}"
        )

class ComputerBuilder(ABC):
    def __init__(self) -> None:
        self._computer = Computer()

    @abstractmethod
    def add_cpu(self) -> None:
        pass

    @abstractmethod
    def add_storage(self) -> None:
        pass

    @abstractmethod
    def add_motherboard(self) -> None:
        pass

    def get_result(self) -> Computer:
        return self._computer

class OfficeComputerBuilder(ComputerBuilder):
    def add_cpu(self) -> None:
        self._computer.cpu = "Intel Core i3"

    def add_storage(self) -> None:
        self._computer.storage = "256 GB SSD"

    def add_motherboard(self) -> None:
        self._computer.motherboard = "Basic motherboard"

class GamingComputerBuilder(ComputerBuilder):
    def add_cpu(self) -> None:
        self._computer.cpu = "Intel Core i7"

    def add_storage(self) -> None:
        self._computer.storage = "1 TB SSD"

    def add_motherboard(self) -> None:
        self._computer.motherboard = "Gaming motherboard"

class WorkstationComputerBuilder(ComputerBuilder):
    def add_cpu(self) -> None:
        self._computer.cpu = "Intel Core i9"

    def add_storage(self) -> None:
        self._computer.storage = "2 TB SSD"

    def add_motherboard(self) -> None:
        self._computer.motherboard = "Professional motherboard"

class ComputerDirector:
    def __init__(self, builder: ComputerBuilder) -> None:
        self.builder = builder

    def build_computer(self) -> Computer:
        self.builder.add_cpu()
        self.builder.add_storage()
        self.builder.add_motherboard()
        return self.builder.get_result()
```

```python
office_builder = OfficeComputerBuilder()
office_director = ComputerDirector(office_builder)
office_computer = office_director.build_computer()

gaming_builder = GamingComputerBuilder()
gaming_director = ComputerDirector(gaming_builder)
gaming_computer = gaming_director.build_computer()

# Build workstation manually using builder
workstation_builder = WorkstationComputerBuilder()
workstation_builder.add_cpu()
workstation_builder.add_storage()
workstation_builder.add_motherboard()
workstation_computer = workstation_builder.get_result()


print("Office computer:")
print(office_computer)

print("\nGaming computer with upgraded CPU:")
print(gaming_computer)

print("\nWorkstation computer:")
print(workstation_computer)
```
