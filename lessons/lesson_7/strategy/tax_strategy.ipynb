```python
from abc import ABC, abstractmethod

class TaxStrategy(ABC):
    @abstractmethod
    def calculate_tax(self, amount: float) -> float:
        pass

class USATaxStrategy(TaxStrategy):
    def calculate_tax(self, amount: float) -> float:
        return amount * 0.10

class EUTaxStrategy(TaxStrategy):
    def calculate_tax(self, amount: float) -> float:
        return amount * 0.20

class NoTaxStrategy(TaxStrategy):
    def calculate_tax(self, amount: float) -> float:
        return 0

class Product:
    def __init__(self, name: str, price: float) -> None:
        if not name:
            raise ValueError("Product name cannot be empty")

        if price < 0:
            raise ValueError("Product price cannot be negative")

        self.name = name
        self.price = price

class Order:
    def __init__(self, products: list[Product], tax_strategy: TaxStrategy) -> None:
        self.products = products
        self.tax_strategy = tax_strategy

    def calculate_total_without_tax(self) -> float:
        return sum(product.price for product in self.products)

    def calculate_tax(self) -> float:
        return self.tax_strategy.calculate_tax(self.calculate_total_without_tax())

    def calculate_total_with_tax(self) -> float:
        return self.calculate_total_without_tax() + self.calculate_tax()

    def print_statistics(self) -> None:
        print(f"Total without tax: {self.calculate_total_without_tax():.2f}")
        print(f"Tax: {self.calculate_tax():.2f}")
        print(f"Total with tax: {self.calculate_total_with_tax():.2f}")
```

```python
products = [
    Product("Laptop", 1000),
    Product("Mouse", 50),
    Product("Keyboard", 100),
]

usa_order = Order(products, USATaxStrategy())
eu_order = Order(products, EUTaxStrategy())
no_tax_order = Order(products, NoTaxStrategy())

print("USA order:")
usa_order.print_statistics()

print("\nEU order:")
eu_order.print_statistics()

print("\nNo tax order:")
no_tax_order.print_statistics()

print("\nPrice difference:")
print(f"USA vs No tax: {usa_order.calculate_total_with_tax() - no_tax_order.calculate_total_with_tax():.2f}")
print(f"EU vs No tax: {eu_order.calculate_total_with_tax() - no_tax_order.calculate_total_with_tax():.2f}")
print(f"EU vs USA: {eu_order.calculate_total_with_tax() - usa_order.calculate_total_with_tax():.2f}")
```
