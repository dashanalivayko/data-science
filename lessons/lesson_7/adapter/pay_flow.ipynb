```python
from abc import ABC, abstractmethod

class PaymentAdapter(ABC):
    @abstractmethod
    def pay(self, amount: float) -> float:
        pass

class CashPayment:
    def pay_with_cash(self, amount: float) -> float:
        print(f"[CashPayment] Received cash payment: {amount:.2f}")
        return amount

class CardPayment:
    def charge_card(self, amount: float) -> float:
        commission = amount * 0.02
        final_amount = amount - commission

        print(f"[CardPayment] Charged card: {amount:.2f}")
        print(f"[CardPayment] Card commission: {commission:.2f}")
        print(f"[CardPayment] Final received amount: {final_amount:.2f}")

        return final_amount

class CashPaymentAdapter(PaymentAdapter):
    def __init__(self, cash_payment: CashPayment) -> None:
        self.cash_payment = cash_payment

    def pay(self, amount: float) -> float:
        print("[Adapter Log] Client called: pay(amount)")
        print("[Adapter Log] CashPaymentAdapter converts: pay(amount) -> pay_with_cash(amount)")

        return self.cash_payment.pay_with_cash(amount)

class CardPaymentAdapter(PaymentAdapter):
    def __init__(self, card_payment: CardPayment) -> None:
        self.card_payment = card_payment

    def pay(self, amount: float) -> float:
        print("[Adapter Log] Client called: pay(amount)")
        print("[Adapter Log] CardPaymentAdapter converts: pay(amount) -> charge_card(amount)")

        return self.card_payment.charge_card(amount)

def checkout(payment_adapter: PaymentAdapter, amount: float) -> None:
    print("\n--- Checkout started ---")

    received_amount = payment_adapter.pay(amount)

    print(f"[Checkout] Received amount: {received_amount:.2f}")
    print("--- Checkout finished ---")
```

```python
cash_payment = CashPayment()
card_payment = CardPayment()

cash_adapter = CashPaymentAdapter(cash_payment)
card_adapter = CardPaymentAdapter(card_payment)

checkout(cash_adapter, 100)
checkout(card_adapter, 100)
```
