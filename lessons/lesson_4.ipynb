```python
# Zadanie 2

def words_count(s):
    s_split = s.split()
    count = len(s_split)
    count_unq = len(set(s_split))

    dict_words = {}

    for word in s_split:
       s_split.get(dict_words[word], 0) += 1

    most_com = max(dict_words.items(), key=lambda item: item[1])
    summ = 0
    for word in s_split:
        summ += len(word)
    avg = round(summ/count, 2)
    return f"Words: {count}, Unique words: {count_unq}, The most common: {most_com}, Average length: {avg}"

s = "Mama Mia Amerikano Cappuchino Mamma Mia Mia"
words_count(s)
```

```python
# Zadanie 3

def game():
    num = random.randint(1,100)
    attempts = 0
    print("Try to guess the number")
    while True:
        guess = int(input("Your number: "))
        attempts += 1
        if guess < num:
            print("Higher")
            continue
        elif guess > num:
            print("Lower")
            continue
        else:
            print(f"You guessed! Your attempts {attempts}")
            break

game()
```

```python
# Zadanie 4
import time

def timer():
    sec = int(input("Write seconds: "))
    while sec >= 0:
        mins = sec // 60
        secs = sec % 60
        print(f"{mins:02d}:{secs:02d}", end="\r")
        time.sleep(1)
        sec -= 1
    print("Time out!")

timer()       
```
