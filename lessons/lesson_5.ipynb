```python

class Money:
    COINS_IN_RUBLE = 100

    def __init__(self, rubles, coins, currency):
        if rubles < 0 or coins < 0:
            raise ValueError("Money amount cannot be negative")

        if not currency:
            raise ValueError("Currency cannot be empty")

        total_coins = rubles * self.COINS_IN_RUBLE + coins

        self.__rubles = total_coins // self.COINS_IN_RUBLE
        self.__coins = total_coins % self.COINS_IN_RUBLE
        self.__currency = currency

    @property
    def rubles(self):
        return self.__rubles

    @property
    def coins(self):
        return self.__coins

    @property
    def currency(self):
        return self.__currency

    def __str__(self):
        return f"{self.__rubles},{self.__coins:02d} {self.__currency}"

    def print_money(self):
        print(self)

    def __to_total_coins(self):
        return self.__rubles * self.COINS_IN_RUBLE + self.__coins

    def __set_from_total_coins(self, total_coins):
        if total_coins < 0:
            raise ValueError("Money amount cannot be negative")

        self.__rubles = total_coins // self.COINS_IN_RUBLE
        self.__coins = total_coins % self.COINS_IN_RUBLE

    def __check_currency(self, other):
        if self.currency != other.currency:
            raise ValueError("Currencies do not match")

    def add(self, rubles, coins, account_from):
        self.__check_currency(account_from)

        amount_total_coins = rubles * self.COINS_IN_RUBLE + coins

        self_total_coins = self.__to_total_coins()
        account_from_total_coins = account_from.__to_total_coins()

        if account_from_total_coins < amount_total_coins:
            print("Недостаточно средств")
            return None

        self.__set_from_total_coins(self_total_coins + amount_total_coins)
        account_from.__set_from_total_coins(account_from_total_coins - amount_total_coins)

        return None

    def minus(self, rubles, coins, account_to):
        self.__check_currency(account_to)

        amount_total_coins = rubles * self.COINS_IN_RUBLE + coins

        self_total_coins = self.__to_total_coins()
        account_to_total_coins = account_to.__to_total_coins()

        if self_total_coins < amount_total_coins:
            print("Недостаточно средств")
            return None
        
        self.__set_from_total_coins(self_total_coins - amount_total_coins)
        account_to.__set_from_total_coins(account_to_total_coins + amount_total_coins)

        return None

    def multiply(self, number):
        total_coins = round(self.__to_total_coins() * number)

        self.__set_from_total_coins(total_coins)

    def divide(self, number):
        if number == 0:
            raise ZeroDivisionError("Cannot divide by zero")

        total_coins = round(self.__to_total_coins() / number)

        self.__set_from_total_coins(total_coins)

    def compare(self, other):
        self.__check_currency(other)

        my_total = self.__to_total_coins()
        other_total = other.__to_total_coins()

        if my_total > other_total:
            return f'Money1 is higher'

        if my_total < other_total:
            return f'Money2 is higher'

        return f'Money1 == Money2'
```

```python
money1 = Money(34, 67, "RUB")
money2 = Money(30, 60, "RUB")

money1.add(10, 45, money2)
money1.print_money()
money2.print_money() 

money1.minus(0, 5, money2)
money1.print_money()
money2.print_money()

money1.multiply(2.5)
print(money1)

money1.divide(2)
print(money1)

money1.compare(money2)
```
