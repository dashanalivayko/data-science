```python
from abc import ABC, abstractmethod

class Vector(ABC):
    @abstractmethod
    def __add__(self, other):
        pass
    def __str__(self):
        pass
    def __mul__(self, num: float):
        pass
    def __sub__(self, other):
        pass
```

```python
class Vector_1d(Vector):
    def __init__(self, arr: list):
        for item in arr:
            if not isinstance(item, float | int) :
                raise ValueError("Только числа !!!")
        self.len = len(arr)
        self.arr = arr

    def __add__(self, other: "Vector_1d") -> "Vector_1d":
        if not isinstance(other, Vector_1d):
            return None
        if self.len != other.len:
            return None
        result = []
        for i in range(self.len):
            result.append(self.arr[i] + other.arr[i])

        return Vector_1d(result)

    def __str__(self):
        return str(self.arr)

    def __sub__(self, other: "Vector_1d") -> "Vector_1d":
        if not isinstance(other, Vector_1d):
            return None
        if self.len != other.len:
            return None
        result = []
        for i in range(self.len):
            result.append(self.arr[i] - other.arr[i])

        return Vector_1d(result)

    def __mul__(self, num: float) -> "Vector_1d":
        return Vector_1d([item * num for item in self.arr])
    
```

```python
class Vector_2d(Vector):
    def __init__(self, arr: list[list]):
        self.arr: list[Vector_1d] = []

        for item in arr:
            if isinstance(item, list) and len(item) == len(arr[0]):
                self.arr.append(Vector_1d(item))
            else:
                raise ValueError("Только листы !!!")
            
        self.row_count = len(arr)
        self.column_count = len(arr[0])

    def __add__(self, other: "Vector_2d") -> "Vector_2d":
        if not isinstance(other, Vector_2d):
            return None
        
        if self.row_count != other.row_count  or self.column_count != other.column_count:
            return None
        
        result = []
        
        for i in range(self.row_count):
            row = []
            for j in range(self.column_count):
                row.append(self.arr[i].arr[j] + other.arr[i].arr[j])
            result.append(row)

        return Vector_2d(result)

    def __str__(self):
        result = []

        for item in self.arr:
            result.append(item.arr)

        return str(result)

    def __sub__(self, other: "Vector_2d") -> "Vector_1d":
        if not isinstance(other, Vector_2d):
            return None
        
        if self.row_count != other.row_count  or self.column_count != other.column_count:
            return None
        
        result = []

        for i in range(self.row_count):
            row = []
            for j in range(self.colum_count):
                row.append(self.arr[i].arr[j] - other.arr[i].arr[j])
            result.append(row)

        return Vector_2d(result)


    def __mul__(self, num: float) -> "Vector_2d":
        return Vector_2d([(item * num).arr for item in self.arr])
    
 
    def transpon(self):
        result = []
        for i in range(self.row_count):
            row = []
            for j in range(self.colum_count):
                row.append(self.arr[j].arr[i] )
            result.append(row)

        return Vector_2d(result)
        
```

```python
class Vector_3d(Vector):
    def __init__(self, arr: list[list[list]]):
        if not isinstance(arr, list) or len(arr) == 0:
            raise ValueError("Нужен непустой 3D список !!!")

        if not isinstance(arr[0], list) or len(arr[0]) == 0:
            raise ValueError("Нужен список матриц !!!")

        if not isinstance(arr[0][0], list) or len(arr[0][0]) == 0:
            raise ValueError("Нужен список строк !!!")

        self.arr = []

        self.depth_count = len(arr)
        self.row_count = len(arr[0])
        self.colum_count = len(arr[0][0])

        for layer in arr:
            if not isinstance(layer, list):
                raise ValueError("Каждый слой должен быть листом !!!")

            if len(layer) != self.row_count:
                raise ValueError("Все слои должны иметь одинаковое количество строк !!!")

            for row in layer:
                if not isinstance(row, list):
                    raise ValueError("Каждая строка должна быть листом !!!")

                if len(row) != self.colum_count:
                    raise ValueError("Все строки должны иметь одинаковую длину !!!")

            self.arr.append(Vector_2d(layer))

    def __add__(self, other: "Vector_3d") -> "Vector_3d":
        if not isinstance(other, Vector_3d):
            return None

        if (
            self.depth_count != other.depth_count
            or self.row_count != other.row_count
            or self.colum_count != other.colum_count
        ):
            return None

        result = []

        for d in range(self.depth_count):
            layer = []

            for i in range(self.row_count):
                row = []

                for j in range(self.colum_count):
                    value = self.arr[d].arr[i].arr[j] + other.arr[d].arr[i].arr[j]
                    row.append(value)

                layer.append(row)

            result.append(layer)

        return Vector_3d(result)

    def __sub__(self, other: "Vector_3d") -> "Vector_3d":
        if not isinstance(other, Vector_3d):
            return None

        if (
            self.depth_count != other.depth_count
            or self.row_count != other.row_count
            or self.colum_count != other.colum_count
        ):
            return None

        result = []

        for d in range(self.depth_count):
            layer = []

            for i in range(self.row_count):
                row = []

                for j in range(self.colum_count):
                    value = self.arr[d].arr[i].arr[j] - other.arr[d].arr[i].arr[j]
                    row.append(value)

                layer.append(row)

            result.append(layer)

        return Vector_3d(result)

    def __mul__(self, num: float) -> "Vector_3d":
        if not isinstance(num, int | float):
            raise ValueError("Умножать можно только на число !!!")


        return Vector_3d([
            [(row * num).arr for row in layer.arr]
            for layer in self.arr
        ])

    def __str__(self):
        result = []

        for layer in self.arr:
            current_layer = []

            for row in layer.arr:
                current_layer.append(row.arr)

            result.append(current_layer)

        return str(result)
```

```python
cube1 = Vector_3d([
    [
        [1, 2],
        [3, 4],
    ],
    [
        [5, 6],
        [7, 8],
    ],
])

cube2 = Vector_3d([
    [
        [10, 20],
        [30, 40],
    ],
    [
        [50, 60],
        [70, 80],
    ],
])

result = cube1 + cube2
print(result)
result = cube2 - cube1
print(result)
result = cube1 * 2
print(result)
```
