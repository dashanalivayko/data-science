```python
# task 1
# string compression

def string_compression(text: str) -> str:
    prev: str = None
    count: int = 0
    res: str = ''

    for char in text:
        if prev:
            if char == prev:
                count += 1
            else:
                res = res + prev + str(count + 1)
                count = 0
            prev = char
        else:
            prev = char

    res += prev + str(count)

    return res

string_compression('qqwwkdjrrrxxxx')
```

```python
# task 2
# almost eaqual string

def is_string_almost_equal(s1: str, s2: str) -> bool:
    i = 0
    j = 0
    errors = 0
    
    if abs(len(s1) - len(s2)) > 1:
        return False

    while i < len(s1) and j < len(s2):
        if s1[i] == s2[j]:
            i += 1
            j += 1
        else:
            errors += 1
            if errors > 1:
                return False

            if len(s1) == len(s2):
                i += 1
                j += 1
            elif len(s1) < len(s2):
                j += 1
            else:
                i += 1

    return True
    
is_string_almost_equal('asdf', 'asdef')
```

```python
# task 3
# longest palindrome

def longest_palindrome(s: str) -> str:
    res = ''

    def search(l, r):
        nonlocal res
        while l >= 0 and r < len(s) and s[l] == s[r]:
            current = s[l:r+1]
            if len(current) > len(res):
                res = current
            l -= 1
            r += 1
    
    for i in range(len(s)):
        search(i, i)
        search(i, i + 1)

    return res

longest_palindrome('cmqwertyytrewqwo')
```

```python
# matrix rotate

def rotate_matrix_180(matrix: list[list]) -> list[list]:
    if not matrix:
        return []
    
    rolled = [row[::-1] for row in matrix[::-1]]
    return rolled
            


matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]


rotate_matrix_180(matrix)
```

```python
# 2

def magic_square(matrix):
    if not matrix:
        return []

    n = len(matrix)

    for row in matrix:
        if len(row) != n:
            return False

    target = sum(matrix[0])

    for row in matrix:
        if sum(row) != target:
            return False

    for j in range(n):
        colum = [row[j] for row in matrix]
        if sum(colum) != target:
            return False

    main_diag = sum([matrix[i][i] for i in range(n)])
    if main_diag != target:
        return False

    second_diag = sum([matrix[n - i - 1][i] for i in range(n)])
    if second_diag != target:
        return False

    return True


matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_2 = [[2, 2, 2], [2, 2, 2], [2, 2, 2]]


magic_square(matrix_2)
```

```python
#3
def roll_slice(matrix , k):
    if not matrix:
        return []
    n = len(matrix)
    outside = []

    for j in range(1,n):
        outside.append(matrix[0][j])
    
    for j in range(1,n):
        outside.append(matrix[j][n-1])

    for j in range(n-2,-1,-1):
        outside.append(matrix[n-1][j])
    
    for j in range(n-2,-1,-1):
        outside.append(matrix[j][0])
    
    rotated = outside[-k:]+outside[:-k] 



    for j in range(1,n):
        matrix[0][j] = rotated.pop(0)

    
    for j in range(1,n):
        matrix[j][n-1] = rotated.pop(0)

    for j in range(n-2,-1,-1):
        matrix[n-1][j] = rotated.pop(0)
    
    for j in range(n-2,-1,-1):
        matrix[j][0] = rotated.pop(0)
    
    return matrix
    
          


matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]


roll_slice(matrix , 2)
```
