```python
# task 1
# string compression

def string_compression(text: str) -> str:
    prev: str = None
    count: int = 0
    res: str = ''

    for char in text:
        if prev:
            if char == prev:
                count += 1
            else:
                res = res + prev + str(count + 1)
                count = 0
            prev = char
        else:
            prev = char

    res += prev + str(count)

    return res

string_compression('qqwwkdjrrrxxxx')
```

```python
# task 2
# almost eaqual string

def is_string_almost_equal(s1: str, s2: str) -> bool:
    i = 0
    j = 0
    errors = 0
    
    if abs(len(s1) - len(s2)) > 1:
        return False

    while i < len(s1) and j < len(s2):
        if s1[i] == s2[j]:
            i += 1
            j += 1
        else:
            errors += 1
            if errors > 1:
                return False

            if len(s1) == len(s2):
                i += 1
                j += 1
            elif len(s1) < len(s2):
                j += 1
            else:
                i += 1

    return True
    
is_string_almost_equal('asdf', 'asdef')
```

```python
# task 3
# longest palindrome

def longest_palindrome(s: str) -> str:
    res = ''

    def search(l, r):
        nonlocal res
        while l >= 0 and r < len(s) and s[l] == s[r]:
            current = s[l:r+1]
            if len(current) > len(res):
                res = current
            l -= 1
            r += 1
    
    for i in range(len(s)):
        search(i, i)
        search(i, i + 1)

    return res

longest_palindrome('cmqwertyytrewqwo')
```
